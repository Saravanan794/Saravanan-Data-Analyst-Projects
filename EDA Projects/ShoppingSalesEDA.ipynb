```python
df["Gender"].value_counts().plot(kind="bar",color="blue")
plt.title("Gender in bars")
plt.show()
```

```python
df["Season"].value_counts().plot(kind="bar")
plt.xlabel("Season")
plt.ylabel("Count")
plt.title("Count of purchhases in each season")
plt.show()
```

```python
df["Category"].value_counts().plot(kind="pie")
plt.title("Pie chart by Category")
plt.show()
```

```python
df["Purchase Amount (USD)"].plot(kind="hist",bins=10)
plt.title("Histogram of purchase amount")
plt.xlabel("USD")
plt.show()
```

```python
df.groupby("Season")["Purchase Amount (USD)"].min().plot(kind="pie")
plt.xlabel("Season")
plt.ylabel("Sum of Purchase Amount (USD)")
plt.show()
```
